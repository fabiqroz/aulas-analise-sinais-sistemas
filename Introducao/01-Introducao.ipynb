## Nota√ß√µes matem√°ticas √∫teis <a class="anchor" id="section_2_1_1"></a>

* Conjuntos ser√£o definidos por nota√ß√£o apropriada (letras mai√∫sculas simpples) e seus elementos s√£o delimitados por chaves ou paranteses

  Ex.: $A = [a_1 ... a_n) = \{a_i \in \mathbb{Z} : a_1 \leq a_i < a_n\} = \{a_1, a_2, ..., a_{n-1}\}$

* Um mapeamento √© um relacionamento envolvendo dois conjuntos que associa cada elemento de um conjunto, denominado dom√≠nio, a um
elemento do outro conjunto, chamado contradom√≠nio. A nota√ß√£o $f:A \rightarrow  B$ denota um mapeamento $f$ cujo dom√≠nio √© o
conjunto $A$ e cujo contradom√≠nio √© o conjunto $B$. Mapeamentos ser√£o sempre definidas por letras min√∫sculas simples.
  
  Ex.: $A=\{1,2,3,4\}$ e $A=\{0,1,2,3\}$ Seja, $f:A \rightarrow  B$, ent√£o:
  $f(t) = \begin{cases}0 \;\; t \in \{1,2\} \\ 1 \;\; t = 5 \\ 2 \;\; t = 3 \end{cases}$

* Uma fun√ß√£o √© um mapeamento em que o dom√≠nio √© um conjunto de natureza cont√≠nua, como n√∫meros reais, n√∫meros complexos ou um subconjunto de qualquer um deles. Na pr√°tica, o contradom√≠nio normalmente √© composto por n√∫meros reais ou n√∫meros complexos. As fun√ß√µes tamb√©m s√£o comumente chamadas de **sinais de tempo cont√≠nuo (CT)**. Aqui, nos concentramos principalmente em
fun√ß√µes de uma √∫nica vari√°vel independente (ou seja, fun√ß√µes unidimensionais). A fun√ß√£o trigonom√©trica seno √© um exemplo de fun√ß√£o:

  $f: \mathbb{R} \rightarrow \mathbb{R}$, tal que $f(t) = \sin(t)$ 
  
* Uma sequ√™ncia √© um mapeamento em que o dom√≠nio √© um conjunto de natureza discreta, como os inteiros ou um subconjunto deles.
Na pr√°tica, o contradom√≠nio normalmente √© composto por n√∫meros reais ou n√∫meros complexos. As sequ√™ncias tamb√©m s√£o comumente
chamados de **sinais de tempo discreto (DT)**. Aqui, nos concentramos principalmente em sequ√™ncias com uma √∫nica vari√°vel independente
(ou seja, sequ√™ncias unidimensionais). A sequ√™ncia $1^2, 2^2, 3^2, 4^2,...$ de quadrados perfeitos √© um exemplo de sequ√™ncia. Isto
associa cada n√∫mero inteiro (estritamente) positivo $n$ ao n√∫mero inteiro $n^2$ . Neste caso, o dom√≠nio e o contradom√≠nio s√£o os positivos inteiros e podemos expressar a sequencia como:

  $f: \mathbb{Z} \rightarrow \mathbb{Z}$, tal que $f[n] = n^2$

### Criando sinais no python <a class="anchor" id="section_2_1_2"></a>

```python
import matplotlib.pyplot as plt
import numpy as np
```

üü£ $x(t) = t^3 - \frac{t^2}{2} - 2t$

```python
# Taxa de amostragem do sinal que iremos gerar (discretiza√ß√£o no tempo)
n = 50000

# Definindo a amostra no tempo
t = np.linspace(0, 10, n)

# Fun√ß√£o que calcula o valor da amplitude do sinal para cada instante de tempo amostrado
def sinal(t):
    return np.power(t,3) - np.power(t,2)/2 - 2*t

# Vizualiza√ß√£o
plt.plot(t, sinal(t))
plt.grid(linestyle='--')
plt.xlabel("t")
plt.ylabel("x(t)")
plt.show()
```

### Sinal Ca√≥tico ‚ù§Ô∏è‚ù§Ô∏è‚ù§Ô∏è‚ù§Ô∏è

```python
def logistic_map(R, x_t, steps):
    x_t_array = np.empty(steps, dtype=np.float64)

    for i in range(steps):
        x_t_array[i] = x_t
        x_t = R*x_t*(1-x_t)

    return x_t_array

t = np.arange(80)
x_t_array = logistic_map(4, 0.2, t.size)
x_t_array_1 = logistic_map(3.99,0.2000000000000001, t.size)

fig, axs = plt.subplots(2)
fig.suptitle(r"Dynamical systems and chaos: Logistic map")
axs[0].plot(t, x_t_array, '-o', markersize = 3, linewidth = 0.5)
axs[0].plot(t, x_t_array_1, '--o', markersize = 3, linewidth = 0.5)
axs[0].set_title('Time Series for a logistic map with two different initial conditions')
axs[1].plot(t, x_t_array-x_t_array_1, '-o', color = 'green', markersize = 3, linewidth = 0.5)
axs[1].set_title('Differences between the two time series')
fig.tight_layout()
plt.show()
```

## Tamanho de um Sinal <a class="anchor" id="section_2_4_1"></a>

O tamanho de qualquer entidade √© um n√∫mero que indica a largura ou o comprimento da entidade. Genericamente falando, a amplitude do sinal varia com o tempo. 

**Como um sinal que existe em um certo intervalo de tempo com amplitude variante pode ser medido por um n√∫mero que ir√° indicar o tamanho ou a for√ßa do sinal?** 

Tal medida deve considerar n√£o apenas a **amplitude** do sinal, mas tamb√©m sua **dura√ß√£o**. 


## Energia de um Sinal <a class="anchor" id="section_2_4_2"></a>

Argumentando desta forma, podemos considerar a √°rea abaixo do sinal $x(t)$ sob um determinado intervalo $[t_1, t_2]$ como uma poss√≠vel medida de seu tamanho, pois a √°rea ir√° considerar n√£o somente a amplitude, mas tamb√©m sua dura√ß√£o.

$\text{Tamanho} = \int_{t_1}^{t_2} x(t)dt$


```python
# Taxa de amostragem do sinal que iremos gerar
n = 1000

# Amostrando o sinal no tempo
t = np.linspace(0, 1, n, endpoint=True)

#Fun√ß√£o que calcula o valor da amplitude do sinal para cada instante de tempo amostrado
sinal = np.power(t,2)


#vizualiza√ß√£o
plt.rcParams["figure.figsize"] = [7.50, 3.50]

plt.fill_between(t, sinal, step="pre", alpha=0.4)
plt.plot(t, sinal, drawstyle="steps")
plt.xlabel('tempo (t)')
plt.ylabel('Amplitude  x(t)')

plt.show()
```

```python
import scipy.integrate as integrate
```

```python
# c√°lculo de integral (tamanho do sinal) via m√©todo num√©rico
tamanho, erro  = integrate.quad(lambda t: t**2, 0, 1)
print(tamanho)
```

Entretanto, esta medida ainda √© defeituosa, pois mesmo para um sinal grande $x(t)$, suas √°reas positivas e negativas podem se cancelar, indicando um sinal de tamanho pequeno:

```python
# Taxa de amostragem do sinal que iremos gerar
n = 1000

# Amostrando o sinal no tempo
t = np.linspace(-10, 10, n, endpoint=False)

#Fun√ß√£o que calcula o valor da amplitude do sinal para cada instante de tempo amostrado
sinal = np.sin(t)

#vizualiza√ß√£o
plt.rcParams["figure.figsize"] = [7.50, 3.50]

plt.fill_between(t, sinal, step="pre", alpha=0.4)
plt.plot(t, sinal, drawstyle="steps")
plt.xlabel('tempo (t)')
plt.ylabel('Amplitude  x(t)')

plt.show()
```

```python
# c√°lculo de integral (tamanho do sinal) via m√©todo num√©rico
tamanho, erro  = integrate.quad(lambda x: np.sin(x), -10, 10)
print(tamanho)
```

Esta dificuldade pode ser corrigida pela defini√ß√£o do tamanho do sinal como a √°rea debaixo de $x^2(t)$ a qual √© sempre positiva. Podemos chamar essa medida de **energia do sinal dissipara em um intervalo de tempo**: $$\text{Tamanho} = E =  \int_{t_1}^{t_2} x^2(t)dt\$$

```python
#Fun√ß√£o que calcula o valor da amplitude do sinal para cada instante de tempo amostrado
sinal = np.power(np.sin(t),2)

#vizualiza√ß√£o
plt.rcParams["figure.figsize"] = [7.50, 3.50]

plt.fill_between(t, sinal, step="pre", alpha=0.4)
plt.plot(t, sinal, drawstyle="steps")
plt.xlabel('tempo (t)')
plt.ylabel('Amplitude  x(t)')

plt.show()
```

```python
# c√°lculo de integral (tamanho do sinal) via m√©todo num√©rico
tamanho, erro  = integrate.quad(lambda x: np.power(np.sin(x),2), -10, 10)
print(tamanho)
```

Essa defini√ß√£o pode ser generalizada para um sinal complexo $x(t)$, sendo dada por $$E = \int_{t_1}^{t_2} |x(t)|^2 dt$$

```python
t = np.linspace(0, 0.05, 1000)

#Fun√ß√£o com componente complexa que calcula o valor da amplitude do sinal para cada instante de tempo amostrado
sinal = np.exp(2j*np.pi*50*t)

#vizualiza√ß√£o
plt.subplot(2,1,1)
plt.fill_between(t, sinal.real, step="pre", alpha=0.4)
plt.plot(t, sinal.real)
plt.xlabel('tempo (t)')
plt.ylabel('Real  x(t)')
plt.title(r'Parte real de $x(t)=e^{j 100 \pi t}$')

plt.subplot(2,1,2)
plt.fill_between(t, sinal.imag, step="pre", alpha=0.4)
plt.plot(t, sinal.imag )
plt.xlabel('tempo (t)')
plt.ylabel('Im  x(t)')
plt.title(r'Parte imagin√°ria de $x(t)=e^{j 100 \pi t}$')
plt.tight_layout()
plt.show()
```

```python
# c√°lculo de integral (tamanho do sinal) via m√©todo num√©rico
energia, erro  = integrate.quad(lambda t: np.power(np.abs(np.exp(2j*np.pi*50*t)),2), 0, 0.05, complex_func=True)
print(energia)
```

## Potencia de um sinal  <a class="anchor" id="section_2_4_3"></a>

Quando a amplitude do sinal x(t) n√£o ‚Üí 0 quando |t|‚Üí‚àû, a **energia do sinal √© infinita** üò∞. 

Nestes. casos, uma medida mais significativa do tamanho do sinal neste caso √© a **energia m√©dia**, se ela existir.  Esta medida √© chamada de **pot√™ncia do sinal**. 

Para um sinal $x(t)$, definimos sua pot√™ncia $P$ por 

$$P = \lim_{T \rightarrow \infty} \int_{-T/2}^{T/2} |x(t)|^2 dt$$

Observe que a pot√™ncia $P$ do sinal √© uma **m√©dia temporal do quadrado da amplitude do sinal**, ou seja, o valor m√©dio quadr√°tico de $x(t)$. 


Geralmente, a m√©dia de uma entidade ao longo de um grande intervalo de tempo aproximando do infinito existe se a entidade for peri√≥dica ou possuir uma regularidade estat√≠stica (para o caso de sinais n√£o-determin√≠sticos).

A pot√™ncia em um determinado intervalo de tempo √©:

$$P = \frac{1}{t_2-t_1}‚à´_{t_1}^{t_2} x^2(t) dt$$


## Sinais de energia e sinais de potencia



*  Um sinal de energia √© um sinal em que a energia √© finita (ex: sinais n√£o peri√≥dicos)
*  Um sinal de pot√™ncia √© um sinal em que a pot√™ncia √© finita e n√£o nula (sinais peri√≥dicos)
*  Ambos os sinais de energia e pot√™ncia s√£o mutuamente exclusivos, ou seja, nenhum sinal pode ser sinal de pot√™ncia e sinal de energia.
* Todos os sinais pr√°ticos t√™m energia finita; portanto, s√£o sinais de energia.
* Na pr√°tica, a gera√ß√£o f√≠sica do sinal de pot√™ncia √© imposs√≠vel, pois requer dura√ß√£o infinita e energia infinita.
* A soma de um sinal de energia e um sinal de pot√™ncia √© um sinal de pot√™ncia.
* Um sinal n√£o √© um sinal de energia nem de pot√™ncia se tanto a energia quanto a pot√™ncia do sinal forem iguais ao infinito.
* Um sinal cuja amplitude √© constante ao longo de uma dura√ß√£o infinita √© um sinal de pot√™ncia.

## Fun√ß√£o Degrau Unit√°rio $u(t)$

Em v√°rias de nossas discuss√µes, os sinais come√ßam em t = 0 (sinais causais). Tais sinais podem ser convenientemente descritos em termos da fun√ß√£o degrau unit√°rio $u(t)$: 

$$u(t) = \begin{cases} 1, & t\geq 0 \\ 0, & t< 0 \end{cases}$$

<img src = "https://drive.google.com/uc?id=1GdiDjLc31NEpLl2Kj1w_Fp0Mf2UDCPbh" width = "600">

Se quisermos um sinal que comece em $t = 0$ (de tal forma que ele possua valor nulo para $t < 0$), precisamos apenas multiplicar o sinal por $u(t)$. 

```python
# Taxa de amostragem do sinal que iremos gerar
n = 1000

# Amostrando o sinal no tempo
t = np.linspace(-10, 10, n, endpoint=False)

#Fun√ß√£o que calcula o valor da amplitude do sinal para cada instante de tempo amostrado
sinal_ori = np.sin(t)
degrau = np.heaviside(t,0)
sinal_final = sinal_ori*degrau

#vizualiza√ß√£o
plt.subplot(3,1,1)
plt.plot(t, sinal_ori)
plt.xlabel('tempo (t)')
plt.ylabel('x(t)')
plt.title(r'Sinal Original $x(t)=\sin{t}$')

plt.subplot(3,1,2)
plt.plot(t, degrau)
plt.xlabel('tempo (t)')
plt.ylabel('u(t)')
plt.title(r'Fun√ß√£o degrau $u(t)$')


plt.subplot(3,1,3)
plt.plot(t, sinal_final)
plt.xlabel('tempo (t)')
plt.ylabel('u(t)')
plt.title(r'Silnal Resultante $u(t)x(t)$')

plt.tight_layout()
plt.show()
```

##  A Fun√ß√£o Impulso Unit√°rio $Œ¥(t)$

A fun√ß√£o impulso unit√°rio $Œ¥(t)$ √© uma das mais importantes fun√ß√µes no estudo de sinais e sistemas.

$$Œ¥(t) = 0 \; \; t \neq 0$$

$$\int_{-‚àû}^{‚àû} Œ¥(t)dt = 1$$

A caracter√≠stica importante da fun√ß√£o impulso unit√°rio n√£o √© sua forma, mas o fato de que sua dura√ß√£o efetiva (largura do pulso) tende para zero enquanto que a sua √°rea permanece unit√°ria. 

### MULTIPLICA√á√ÉO DE UMA FUN√á√ÉO POR UM IMPULSO

Vamos considerar o que acontece quando multiplicamos o impulso unit√°rio $Œ¥(t)$ por uma fun√ß√£o $œÜ(t)$ que sabemos ser cont√≠nua para $t = 0$. Como o impulso possui valor n√£o nulo apenas para $t = 0$, e o $valor œÜ(t)$ para $t = 0$
√© $œÜ(0)$, obtemos

$$œÜ(t)Œ¥(t) = œÜ(0)Œ¥(t)$$

```python
from scipy import signal

signal.unit_impulse(8)

imp = signal.unit_impulse(1000, 'mid')


# Taxa de amostragem do sinal que iremos gerar
n = 1000

# Amostrando o sinal no tempo
t = np.linspace(-1, 1, n, endpoint=False)
sinal_ori = 2*t**5+2
sinal_final = sinal_ori*imp

# import numpy as np
# import matplotlib.pyplot as pltplt.plot(t, imp)
plt.plot(t, sinal_ori)
plt.plot(t, sinal_final)
# plt.margins(0.1, 0.1)
# plt.xlabel('Time [samples]')
# plt.ylabel('Amplitude')
# plt.grid(True)
plt.show()
```

## Fun√ß√£o Exponencial $e^{st}$

Outra importante fun√ß√£o na √°rea de sinais e sistemas √© o sinal exponencial $e^{st}$, onde s √©, geralmente, um n√∫mero complexo, dado por

$s = œÉ +jw$

Logo, 

$e^{st} = e^{(œÉ +jw)t} = e^{œÉt} e^{jwt} = e^{œÉt}(\cos wt + j \sin wt)$

```python
# Taxa de amostragem do sinal que iremos gerar
n = 1000

# Amostrando o sinal no tempo
t = np.linspace(-10, 10, n, endpoint=False)

sigma = 1
w = 5
sinal_exp =  np.exp((sigma + w*1j)*t)
plt.plot(np.exp(sigma*t))
plt.plot(sinal_exp)
plt.show()
```
