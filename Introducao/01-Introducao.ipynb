## Sinais contínuos e discretos no tempo <a class="anchor" id="section_2_1"></a>

No caso dos sinais de tempo contínuo, a **variável independente é contínua** e, portanto, esses sinais são
definidos em um conjunto contínuo de valores da variável independente. Em contrapartida, os sinais de tempo
discreto são definidos somente em instantes discretos, ou seja, a **variável independente assume apenas um conjunto discreto de valores**.

Exemplos:

1. Um sinal de fala em função do tempo e a pressão atmosférica em função da altitude são exemplos de sinais de tempo contínuo. 
2. O índice semanal Dow-Jones da Bolsa de Valores de Nova York é um exemplo de sinal de tempo discreto

Um sinal de tempo discreto $x[n]$ pode representar um fenômeno para o qual a variável independente é inerentemente discreta. Por outro lado, **uma classe muito importante de sinais de tempo discreto decorre da amostragem de sinais de tempo contínuo**. Nesse caso, o sinal de tempo discreto $x[n]$ representa amostras sucessivas de um fenômeno para o qual a variável independente é contínua.

Um sinal **discreto no tempo**  pode ser *inerentemente discreto* ou corresponder a uma versão *amostrada* de um sinal de tempo contínuo. Um exemplo do primeiro seria um sinal correspondente ao mercado de ações Dow Jones Industrial Average
índice (que é definido apenas em intervalos diários), enquanto um exemplo do último seria a versão amostrada de um (tempo contínuo) sinal de fala.

## Sinais Periódicos e não-periódicos <a class="anchor" id="section_2_3"></a>

Um sinal $x(t)$ é dito periódico se para alguma constante positiva $T$ $$x(t) = x(t+T) \;\; \text{para todo t} $$

Em outras palavras, um sinal periódico tem a propriedade de não se modificar pelo deslocamento no tempo de $T$.

* O menor valor de $T$ que satisfaça a equação é o **período fundamental** do sinal $x(t)$

<img src = "https://drive.google.com/uc?id=1Zhew2Nh33ulnv1CYPnL6fpp_5ZuIrxOc" width = "700">





```python
def periodic_function(func, period):
    return lambda x: func(((x ) % period )  )

# Taxa de amostragem do sinal que iremos gerar
n = 100
T = 10

# # Amostrando o sinal no tempo
a = np.linspace(0, 30, n, endpoint=False)

sinal = periodic_function(lambda x: x, T)

plt.plot(a,sinal(a))
plt.show()
```

Um sinal que não começar antes de $t = 0$ é um **sinal causal**. Em outras palavras, x(t) é um sinal causal se:

$$x(t) = 0 \;\;\; t < 0$$

* Sinais Causais:

<img src = "https://drive.google.com/uc?id=1p3IU03CJ6s8WxbX3SVw4Np1qgVrD7tbT" width = "600">

* Sinais Não-Causais:

<img src = "https://drive.google.com/uc?id=1dZ5HnL4xY0iJ9x5bc_5velqzE2H5uBQr" width = "600">



```python
# Gera sinal causal com frequência de F Hz
def sinal_senoidal(periodo, frequencia, tempo):
    sinal = np.sin(periodo*frequencia*tempo)
    return sinal

# Taxa de amostragem do sinal que iremos gerar
n = 1000

# Amostrando o sinal no tempo
x = np.linspace(0, 5, n, endpoint=False)

# Definindo o período e frequencia do sinal
p = 4*np.pi
f = 1

# Vizualização
plt.plot(x, sinal_senoidal(p,f,x))
plt.show()
```

## Sinais determinísticos e estocásticos <a class="anchor" id="section_2_5"></a>

Um sinal cuja descrição física é completamente conhecida, seja na forma matemática ou na forma gráfica é um sinal determinístico. Um sinal cujos valores não podem ser preditos precisamente, mas são conhecidos apenas em termos de uma descrição probabilística, tal como o valor médio ou valor médio quadrático, são sinais aleatórios. Neste livro, trabalharemos exclusivamente com sinais determinísticos. Os sinais aleatórios estão além do escopo dos nosssos estudos.

```python
# Taxa de amostragem do sinal que iremos gerar
n = 1000

# Amostrando o sinal no tempo
x = np.linspace(1, 100, n, endpoint=False)

#Função que calcula o valor da amplitude do sinal para cada instante de tempo amostrado
def sinal_aleatorio(x):
    return  np.random.normal(scale=0.8, size=len(x))

# Vizualização
plt.plot(x, sinal_aleatorio(x))
plt.show()
```

# Energia e potência de um sinal <a class="anchor" id="section_2_4"></a>

## Tamanho de um Sinal <a class="anchor" id="section_2_4_1"></a>

O tamanho de qualquer entidade é um número que indica a largura ou o comprimento da entidade. Genericamente falando, a amplitude do sinal varia com o tempo. 

**Como um sinal que existe em um certo intervalo de tempo com amplitude variante pode ser medido por um número que irá indicar o tamanho ou a força do sinal?** 

Tal medida deve considerar não apenas a **amplitude** do sinal, mas também sua **duração**. 


## Energia de um Sinal <a class="anchor" id="section_2_4_2"></a>

Argumentando desta forma, podemos considerar a área abaixo do sinal $x(t)$ sob um determinado intervalo $[t_1, t_2]$ como uma possível medida de seu tamanho, pois a área irá considerar não somente a amplitude, mas também sua duração.

$$\text{Tamanho} = \int_{t_1}^{t_2} x(t)dt$$ 


```python
# Taxa de amostragem do sinal que iremos gerar
n = 1000

# Amostrando o sinal no tempo
t = np.linspace(0, 1, n, endpoint=True)

#Função que calcula o valor da amplitude do sinal para cada instante de tempo amostrado
sinal = np.power(t,2)


#vizualização
plt.rcParams["figure.figsize"] = [7.50, 3.50]

plt.fill_between(t, sinal, step="pre", alpha=0.4)
plt.plot(t, sinal, drawstyle="steps")
plt.xlabel('tempo (t)')
plt.ylabel('Amplitude  x(t)')

plt.show()
```

```python
import scipy.integrate as integrate
```

```python
# cálculo de integral (tamanho do sinal) via método numérico
tamanho, erro  = integrate.quad(lambda t: t**2, 0, 1)
print(tamanho)
```

Entretanto, esta medida ainda é defeituosa, pois mesmo para um sinal grande $x(t)$, suas áreas positivas e negativas podem se cancelar, indicando um sinal de tamanho pequeno:

```python
# Taxa de amostragem do sinal que iremos gerar
n = 1000

# Amostrando o sinal no tempo
t = np.linspace(-10, 10, n, endpoint=False)

#Função que calcula o valor da amplitude do sinal para cada instante de tempo amostrado
sinal = np.sin(t)

#vizualização
plt.rcParams["figure.figsize"] = [7.50, 3.50]

plt.fill_between(t, sinal, step="pre", alpha=0.4)
plt.plot(t, sinal, drawstyle="steps")
plt.xlabel('tempo (t)')
plt.ylabel('Amplitude  x(t)')

plt.show()
```

```python
# cálculo de integral (tamanho do sinal) via método numérico
tamanho, erro  = integrate.quad(lambda x: np.sin(x), -10, 10)
print(tamanho)
```

Esta dificuldade pode ser corrigida pela definição do tamanho do sinal como a área debaixo de $x^2(t)$ a qual é sempre positiva. Podemos chamar essa medida de **energia do sinal dissipara em um intervalo de tempo**: $$\text{Tamanho} = E =  \int_{t_1}^{t_2} x^2(t)dt\$$

```python
#Função que calcula o valor da amplitude do sinal para cada instante de tempo amostrado
sinal = np.power(np.sin(t),2)

#vizualização
plt.rcParams["figure.figsize"] = [7.50, 3.50]

plt.fill_between(t, sinal, step="pre", alpha=0.4)
plt.plot(t, sinal, drawstyle="steps")
plt.xlabel('tempo (t)')
plt.ylabel('Amplitude  x(t)')

plt.show()
```

```python
# cálculo de integral (tamanho do sinal) via método numérico
tamanho, erro  = integrate.quad(lambda x: np.power(np.sin(x),2), -10, 10)
print(tamanho)
```

Essa definição pode ser generalizada para um sinal complexo $x(t)$, sendo dada por $$E = \int_{t_1}^{t_2} |x(t)|^2 dt$$

```python
t = np.linspace(0, 0.05, 1000)

#Função com componente complexa que calcula o valor da amplitude do sinal para cada instante de tempo amostrado
sinal = np.exp(2j*np.pi*50*t)

#vizualização
plt.subplot(2,1,1)
plt.fill_between(t, sinal.real, step="pre", alpha=0.4)
plt.plot(t, sinal.real)
plt.xlabel('tempo (t)')
plt.ylabel('Real  x(t)')
plt.title(r'Parte real de $x(t)=e^{j 100 \pi t}$')

plt.subplot(2,1,2)
plt.fill_between(t, sinal.imag, step="pre", alpha=0.4)
plt.plot(t, sinal.imag )
plt.xlabel('tempo (t)')
plt.ylabel('Im  x(t)')
plt.title(r'Parte imaginária de $x(t)=e^{j 100 \pi t}$')
plt.tight_layout()
plt.show()
```

```python
# cálculo de integral (tamanho do sinal) via método numérico
energia, erro  = integrate.quad(lambda t: np.power(np.abs(np.exp(2j*np.pi*50*t)),2), 0, 0.05, complex_func=True)
print(energia)
```

Uma condição necessária para que a energia seja finita é que 

$$x(t) → 0\;\; \text{quando}\;\; |t|→ ∞$$.

## Potencia de um sinal  <a class="anchor" id="section_2_4_3"></a>

Quando a amplitude do sinal x(t) não → 0 quando |t|→∞, a **energia do sinal é infinita** 😰. 

Nestes. casos, uma medida mais significativa do tamanho do sinal neste caso é a **energia média**, se ela existir.  Esta medida é chamada de **potência do sinal**. 

Para um sinal $x(t)$, definimos sua potência $P$ por 

$$P = \lim_{T \rightarrow \infty} \int_{-T/2}^{T/2} |x(t)|^2 dt$$

Observe que a potência $P$ do sinal é uma **média temporal do quadrado da amplitude do sinal**, ou seja, o valor médio quadrático de $x(t)$. 


Geralmente, a média de uma entidade ao longo de um grande intervalo de tempo aproximando do infinito existe se a entidade for periódica ou possuir uma regularidade estatística (para o caso de sinais não-determinísticos).

A potência em um determinado intervalo de tempo é:

$$P = \frac{1}{t_2-t_1}∫_{t_1}^{t_2} x^2(t) dt$$


## Sinais de energia e sinais de potencia



*  Um sinal de energia é um sinal em que a energia é finita (ex: sinais não periódicos)
*  Um sinal de potência é um sinal em que a potência é finita e não nula (sinais periódicos)
*  Ambos os sinais de energia e potência são mutuamente exclusivos, ou seja, nenhum sinal pode ser sinal de potência e sinal de energia.
* Todos os sinais práticos têm energia finita; portanto, são sinais de energia.
* Na prática, a geração física do sinal de potência é impossível, pois requer duração infinita e energia infinita.
* A soma de um sinal de energia e um sinal de potência é um sinal de potência.
* Um sinal não é um sinal de energia nem de potência se tanto a energia quanto a potência do sinal forem iguais ao infinito.
* Um sinal cuja amplitude é constante ao longo de uma duração infinita é um sinal de potência.

### Observando a energia e potência de um sinal que representa um fenômeno físico:

Em muitas aplicações, os sinais que consideramos estão diretamente relacionados a **quantidades físicas** e a partir deles pode-se extrair a **potência ou energia de um sistema físico**.  

Em circuitos elétricos, os sinais podem representar corrente ou tensão. Considere uma tensão $v(t)$ aplicada através de uma resistência $R$ e $i(t)$ é a corrente que flui através dela, conforme mostrado na figura.

<img src = "https://drive.google.com/uc?id=1fa_j0QueZQWt0iFDZ4-KHP0Dtj4fiohN" width = "500">

A potência instantânea na resistência $R$ é dada por:

$$p(t) = v(t)i(t)$$

Pela lei de Ohm:


$$p(t) = v(t)i(t) = v(t)\frac{v(t)}{R} = \frac{v^2(t)}{R}$$

Também

$$𝑝(𝑡) = 𝑖(𝑡)𝑅 ∙ 𝑖(𝑡) = 𝑖^2(𝑡)𝑅$$

<!-- A energia total dissipada no intervalo de tempo $t_1\le t\le t_2$ é 
$$\frac{1}{R} \int_{t_1}^{t_2}v^2(t)dt$$ -->

Quando os valores da resistência $R = 1Ω$, então a potência dissipada nela é conhecida como potência normalizada. Por isso

$$p(t) = v^2(t)$$

Se $v(t)$ ou $i(t)$ é denotado por um sinal de tempo contínuo $x(t)$, então a potência instantânea é igual ao quadrado da amplitude do sinal, ou seja,

$$p(t) = |x(t)|^2$$

Portanto, a potência média ou potência normalizada de um sinal de tempo contínuo $x(t)$ é dada por,

$$P=lim_{T→∞}\frac{1}{T}∫_{−(T/2)}^{T/2}|x(t)|^2dt \;\;\text{Watts}$$

A energia total ou energia normalizada de um sinal de tempo contínuo é definida como

$$E=lim_{T→∞}∫_{−(T/2)}^{T/2}|x(t)|^2dt \;\;\text{Joules}$$

## Função Degrau Unitário $u(t)$

Em várias de nossas discussões, os sinais começam em t = 0 (sinais causais). Tais sinais podem ser convenientemente descritos em termos da função degrau unitário $u(t)$: 

$$u(t) = \begin{cases} 1, & t\geq 0 \\ 0, & t< 0 \end{cases}$$

<img src = "https://drive.google.com/uc?id=1GdiDjLc31NEpLl2Kj1w_Fp0Mf2UDCPbh" width = "600">

Se quisermos um sinal que comece em $t = 0$ (de tal forma que ele possua valor nulo para $t < 0$), precisamos apenas multiplicar o sinal por $u(t)$. 

```python
# Taxa de amostragem do sinal que iremos gerar
n = 1000

# Amostrando o sinal no tempo
t = np.linspace(-10, 10, n, endpoint=False)

#Função que calcula o valor da amplitude do sinal para cada instante de tempo amostrado
sinal_ori = np.sin(t)
degrau = np.heaviside(t,0)
sinal_final = sinal_ori*degrau

#vizualização
plt.subplot(3,1,1)
plt.plot(t, sinal_ori)
plt.xlabel('tempo (t)')
plt.ylabel('x(t)')
plt.title(r'Sinal Original $x(t)=\sin{t}$')

plt.subplot(3,1,2)
plt.plot(t, degrau)
plt.xlabel('tempo (t)')
plt.ylabel('u(t)')
plt.title(r'Função degrau $u(t)$')


plt.subplot(3,1,3)
plt.plot(t, sinal_final)
plt.xlabel('tempo (t)')
plt.ylabel('u(t)')
plt.title(r'Silnal Resultante $u(t)x(t)$')

plt.tight_layout()
plt.show()
```

##  A Função Impulso Unitário $δ(t)$

A função impulso unitário $δ(t)$ é uma das mais importantes funções no estudo de sinais e sistemas.

$$δ(t) = 0 \; \; t \neq 0$$

$$\int_{-∞}^{∞} δ(t)dt = 1$$

<img src = "https://drive.google.com/uc?id=17iddKQiyjCpZwUsxsbLRGyziYINpmJPv" width = "400">


A característica importante da função impulso unitário não é sua forma, mas o fato de que sua duração efetiva (largura do pulso) tende para zero enquanto que a sua área permanece unitária. 

### MULTIPLICAÇÃO DE UMA FUNÇÃO POR UM IMPULSO

Vamos considerar o que acontece quando multiplicamos o impulso unitário $δ(t)$ por uma função $φ(t)$ que sabemos ser contínua para $t = 0$. Como o impulso possui valor não nulo apenas para $t = 0$, e o $valor φ(t)$ para $t = 0$
é $φ(0)$, obtemos

$$φ(t)δ(t) = φ(0)δ(t)$$

```python
from scipy import signal

signal.unit_impulse(8)

imp = signal.unit_impulse(1000, 'mid')


# Taxa de amostragem do sinal que iremos gerar
n = 1000

# Amostrando o sinal no tempo
t = np.linspace(-1, 1, n, endpoint=False)
sinal_ori = 2*t**5+2
sinal_final = sinal_ori*imp

# import numpy as np
# import matplotlib.pyplot as pltplt.plot(t, imp)
plt.plot(t, sinal_ori)
plt.plot(t, sinal_final)
# plt.margins(0.1, 0.1)
# plt.xlabel('Time [samples]')
# plt.ylabel('Amplitude')
# plt.grid(True)
plt.show()
```

## Função Exponencial $e^{st}$

Outra importante função na área de sinais e sistemas é o sinal exponencial $e^{st}$, onde s é, geralmente, um número complexo, dado por

$s = σ +jw$

Logo, 

$e^{st} = e^{(σ +jw)t} = e^{σt} e^{jwt} = e^{σt}(\cos wt + j \sin wt)$

```python
# Taxa de amostragem do sinal que iremos gerar
n = 1000

# Amostrando o sinal no tempo
t = np.linspace(-10, 10, n, endpoint=False)

sigma = 1
w = 5
sinal_exp =  np.exp((sigma + w*1j)*t)
plt.plot(np.exp(sigma*t))
plt.plot(sinal_exp)
plt.show()
```
