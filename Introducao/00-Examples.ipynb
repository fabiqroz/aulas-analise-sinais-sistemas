# Exemplo 1: Transladando e rotacionando vetores

1. Especifique dois pontos $P_0 = (0,0)$ e $P_1 = (2,1)$.
2. Escreva uma função em python que receba um vetor e translade os dados pela adição de x unidades à primeira coordenada e y unidade à segunda coordenada;
3. Plote o vetor original criado a partir de $P_0$ e $P_1$ e também plote o vetor transladado.

Estenda a solução tão anterior  escrevendo uma função em python  que:

1. recebe um vetor e um ângulo como parametro de entrada
2. rotaciona o vetor por 25º 
3. retorna o vetor rotacionado

Abaixo segue a transformação da rotação de um ponto no espaço bidimensional: 


<img src="../figs/matriz_rotacao.png" alt="sample" width="300">

Após implementação da função, plote o vetor definido por $P_0$ e $P_1$ e, no mesmo gráfico também plote um vetor (de diferente cor) referente ao vetor rotacionado.

```python
import numpy as np

def translada_vetor(p0, p1, x, y):
    # Cria dois arrays zerados para representar o vetor transladado
    p0_trans = np.zeros(p0.shape)
    p1_trans = np.empty(p1.shape)

    # translada as coordenadas x de ambos os pontos
    p0_trans[0] = p0[0] + x
    p1_trans[0] = p1[0] + x

    # translada as coordenadas y de ambos os pontos
    p0_trans[1] = p0[1] + y
    p1_trans[1] = p1[1] + y

    return p0_trans, p1_trans

def rotaciona_vetor(p0, p1, d_ang):
    # Cria uma matriz de zeros de tamanho 2x2 e dados reais
    matriz_rot = np.zeros((2,2))
    
    # Converte angulo para radianos
    ang_rad = np.deg2rad(d_ang)

    matriz_rot = np.array([[np.cos(ang_rad), -np.sin(ang_rad)],
                        [np.sin(ang_rad), np.cos(ang_rad)]])

    # rotaciona as coordenadas de ambos os pontos
    p0_rot = np.matmul(p0, matriz_rot)
    p1_rot = np.matmul(p1, matriz_rot)

    return p0_rot, p1_rot
```

```python
p0_ori = np.array([0, 0])
p1_ori = np.array([1, 2])

p0_trans, p1_trans = translada_vetor(p0_ori, p1_ori, 0, 1)
p0_rot, p1_rot = rotaciona_vetor(p0_ori,p1_ori,45) 
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(1, 2)
fig.suptitle('Transformação sobre vetores')
ax[0].plot([p0_ori[0], p1_ori[0]], [p0_ori[1], p1_ori[1]], c='blue', marker='o')
ax[0].plot([p0_trans[0], p1_trans[0]], [p0_trans[1], p1_trans[1]], c='green', marker='o')
ax[0].set_title('Translação')
ax[0].grid(linestyle='--')

ax[1].plot([p0_ori[0], p1_ori[0]], [p0_ori[1], p1_ori[1]], c='blue', marker='o')
ax[1].plot([p0_rot[0], p1_rot[0]], [p0_rot[1], p1_rot[1]], c='green', marker='o')
ax[1].set_title('Rotação')
ax[1].grid(linestyle='--')
plt.show() 
```

# Exemplo 3: Criando um data frame

```python
import pandas as pd

# Definindo  time stamps de 0 to 10 segundos a uma frequencia de 10 hz
time = np.linspace(0, 10, 100)  # 100 amostras a 10 Hz

# Calculo dos valores de seno e cosseno
y_vals = np.sin(2 * np.pi * 1.5 * time)
z_vals = np.cos(2 * np.pi * 1.5 * time)

# Criando o dataframe a partir de um dicionário dos dados acima definidos
data = {'Time': time, 'YVals': y_vals, 'ZVals': z_vals} # dicionário
df = pd.DataFrame(data) #dataframe
```
